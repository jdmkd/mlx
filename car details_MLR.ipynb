```python
import pandas as pd
from sklearn import linear_model
```

```python
df=pd.read_csv("car details.csv")
```

```python
df
```

```python
x=df[['Selling_Price','Present_Price','Kms_Driven']]
x
```

```python
y=df[['Owner']]
y
```

```python
model.fit(x,y)
```

```python
model.intercept_
```

```python
model.coef_
```
