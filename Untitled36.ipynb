```python
import pandas as pd
from sklearn import linear_model
```

```python
df=pd.read_csv("Dry_Bean_Dataset.csv")
df
```

```python
df_linear=df[['Area','Perimeter']]
df_linear
```

```python
linear=df_linear.drop('Area',axis=1)
linear
```

```python
model=linear_model.LinearRegression()
```

```python
model.fit(linear,df_linear.Area)
```

```python
model.predict([[15]])
```
