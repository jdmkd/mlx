```python
import pandas as pd
```

```python
df=pd.read_csv("playsheet_dataset.csv")
```

```python
df
```

```python
inputs = df.drop('Play',axis=1)
target = df['Play']
```

```python
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
#imoprt library
```

```python
Numerics=LabelEncoder() 
#calling  LabelEncoder
```

```python
inputs['Outlook_n']=Numerics.fit_transform(inputs['Outlook'])
inputs['Temp_n']=Numerics.fit_transform(inputs['Temp'])
inputs['Humidity_n']=Numerics.fit_transform(inputs['Humidity'])
inputs['Windy_n']=Numerics.fit_transform(inputs['Windy'])
inputs
```

```python
inputs_n=inputs.drop(['Outlook','Temp','Humidity','Windy'],axis=1)
inputs_n
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
model.fit(inputs_n,target)
```

```python
model.predict([[2,2,1,1]])
```

```python
model.score(inputs_n,target)*100
```
