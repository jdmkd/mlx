```python
import pandas as pd
```

```python
df=pd.read_csv("NB_loan_data_set.csv")
```

```python
df
```

```python
df.dropna(inplace=True)
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
numerics=LabelEncoder() 
```

```python
df['Gender_new']=numerics.fit_transform(df['Gender'])
df['Married_new']=numerics.fit_transform(df['Married'])
df['Education_new']=numerics.fit_transform(df['Education'])
df['Self_Employed_new']=numerics.fit_transform(df['Self_Employed'])
df['Property_Area_new']=numerics.fit_transform(df['Property_Area'])
df['Loan_Status_new']=numerics.fit_transform(df['Loan_Status'])
```
