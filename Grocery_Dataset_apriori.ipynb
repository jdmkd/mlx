```python
import pandas as pd
from apyori import apriori
```

```python
df=pd.read_csv("Grocery_Dataset.csv")
df
```

```python
records=[]
for i in range(0,1000):
        records.append([str(df.values[i,j]) for j in range(0,7)])
```

```python
association_rules=apriori(records,min_support=0.20,min_confidence=0.4,min_lift=1.8,min_length=2)
```

```python
association_results=list(association_rules)
```

```python
association_results
```

```python
df_results=pd.DataFrame(association_results)
df_results
```
