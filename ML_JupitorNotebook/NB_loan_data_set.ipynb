```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df_drop = df.dropna(how='any')
```

```python
Numerics=LabelEncoder()
```

```python
df_drop['Gender_new']=Numerics.fit_transform(df_drop['Gender'])
df_drop['Married_new']=Numerics.fit_transform(df_drop['Married'])
df_drop['Dependents_new']=Numerics.fit_transform(df_drop['Dependents'])
df_drop['Education_new']=Numerics.fit_transform(df_drop['Education'])
df_drop['Self_Employed_new']=Numerics.fit_transform(df_drop['Self_Employed'])
df_drop['Credit_History_new']=Numerics.fit_transform(df_drop['Credit_History'])
df_drop['Property_Area_new']=Numerics.fit_transform(df_drop['Property_Area'])
df_drop['Loan_Status_new']=Numerics.fit_transform(df_drop['Loan_Status'])
df_drop
```

```python
df=df_drop.drop(['Gender','Married','Dependents','Education','Self_Employed','Credit_History','Property_Area','Loan_Status'],axis=1)
```
