```python
import pandas as pd
df=pd.read_csv(r"Dry_Bean_Dataset.csv")
df
```

```python
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df_linear_regression = df[['Area','Perimeter']]
df_linear_regression
```

```python
plt.xlabel("Area")
plt.ylabel("Perimeter")
plt.scatter(df_linear_regression.Area,df_linear_regression.Perimeter,color="blue", marker=".")
```

```python
new_df=df_linear_regression.drop('Perimeter',axis="columns")
new_df
```

```python
model=linear_model.LinearRegression()
model.fit(new_df,df_linear_regression.Perimeter)
```

```python
model.predict([[250]])
```
