```python
import pandas as pd
df = pd.read_csv(r"Dry_Bean_Dataset.csv")
df
```

```python
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df_liner_regrasion = df[['Area','Perimeter']]
df_liner_regrasion
```

```python
plt.xlabel("area")
plt.ylabel("Perimeter")
plt.scatter(df_liner_regrasion.Area,df_liner_regrasion.Perimeter,color='black',marker='*')
```

```python
liner_model.predict([[5000]])
```

```python
df_KNN = df.dropna(axis=0)
df_KNN
```

```python
Knn_x = df_KNN.iloc[:,[0,1,4,5]].values
Knn_y = df_KNN.iloc[:,16].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(Knn_x, Knn_y, test_size=0.6, random_state=20)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn_model = KNeighborsClassifier(n_neighbors=4)
knn_model.fit(x_train,y_train)
```

```python
knn_y_pred = knn_model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, knn_y_pred)*100
```

```python
new_data_knn = [['41487','759.552','1.688753','0.805826']]
```

```python
knn_new_prediction = knn_model.predict(new_data_knn)
knn_new_prediction
```

```python
#Decision Tree Algoritham

df_decision_tree = df[['Area','Perimeter', 'ConvexArea']]
df_decision_tree
```

```python
x=df_decision_tree.iloc[:,[0,1]].values
x
```

```python
y=df_decision_tree.iloc[:,2].values
y
```

```python
from sklearn.model_selection import train_test_split 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=25)
```

```python
x_train
```

```python
from sklearn.preprocessing import StandardScaler
st_x=StandardScaler()
x_train=st_x.fit_transform(x_train)
x_test=st_x.transform(x_test)
```

```python
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
model=DecisionTreeClassifier(criterion='entropy',random_state=0)
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_test) 
```

```python
cm
```

```python
from sklearn import metrics
metrics.accuracy_score(y_test,y_pred)*100
```
